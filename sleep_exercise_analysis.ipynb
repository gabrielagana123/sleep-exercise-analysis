```python
import pandas as pd

# Load the dataset
df = pd.read_csv('sleep_exercise.csv')

# Display the first few rows
df.head()
```

```python
# Get summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotting the distribution of hours of sleep
sns.histplot(df['Hours_of_Sleep'], kde=True)
plt.title('Distribution of Hours of Sleep')
plt.xlabel('Hours of Sleep')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the distribution of exercise performance
sns.histplot(df['Exercise_Performance'], kde=True)
plt.title('Distribution of Exercise Performance')
plt.xlabel('Exercise Performance')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot to see the relationship between sleep and exercise performance
sns.scatterplot(data=df, x='Hours_of_Sleep', y='Exercise_Performance')
plt.title('Sleep vs. Exercise Performance')
plt.xlabel('Hours of Sleep')
plt.ylabel('Exercise Performance')
plt.show()
```
